```python
def fill_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37.12623188405797
        elif  Pclass==2:
            return 29.549492753623188
        elif Pclass==3:
            return 25.839423076923076
    else:
        return Age
    
```
