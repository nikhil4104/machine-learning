```python
import numpy as np
from sklearn import preprocessing
training=np.loadtxt(r'C:\Users\CW\Downloads\training_boston_x_y_train.csv',delimiter=',')

def step_gradient(X, Y, m, learning_rate):
    slopes=np.zeros(len(X[0]))
    for i in range(len(X)):
        x=X[i]
        y=Y[i]
        for j in range(len(x)):
            slopes[j]+=(-2/len(X))*(y-sum(m*x))*x[j]
    new_m=m-(learning_rate*slopes)
    return new_m

def cost(m, x, y):
    cost=0
    for i in range(len(x)):
        cost+=(1/len(x))*((y[i]-sum(m*x[i]))**2)
    print (cost)
values=[0]*(2000)
values.append(cost)
def gd(x, y, learning_rate, iterations):
    m=np.zeros(len(x[0]))
    for i in range(iterations):
        m=step_gradient(x, y, m, learning_rate)
        print( "cost=", end=' ')
        cost(m, x, y)
    return m

def gradient_descent(x, y):
    iterations=200
    learning_rate=0.03
    x=np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    m=gd(x, y, learning_rate, iterations)
    return m
# In [3]:
x=training[:, :-1]
y=training[:, -1]

#adding squared values of each column

sq=[]
for i in range(len(x[0])):
    for j in range(i, len(x[0])):
        sq.append(x[:, i]*x[:, j])
sq=np.array(sq)
for i in sq:
    x=np.append(x, i.reshape(-1, 1), axis=1)
    
#scaling

scaler=preprocessing.StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
m=gradient_descent(x, y)
# v=cost(m,x,y)
# print(v)
# print(values)
```
